#DATA PROCESSING

```python
#select only relevant columns

cols=['movie_id','title','genres','keywords','overview','cast','crew']
movies_df=movies_df[cols]
```

```python
movies_df.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(x):
  l=[]
  for diction in ast.literal_eval(x):
    l.append(diction['name'])
  return l
```

```python
movies_df['genres']=movies_df['genres'].apply(convert)
```

```python
movies_df['keywords']=movies_df['keywords'].apply(convert)
```

```python
movies_df['overview'][0]
```

```python
def convert_cast(x):
  l=[]
  counter=0
  for item in ast.literal_eval(x):
    if counter!=3:
      l.append(item['name'])
      counter+=1
    else:
      break
  return l

```

```python
movies_df['cast']=movies_df['cast'].apply(convert_cast)
```

```python
movies_df['crew'][0]
```

```python
movies_df['crew'][1]
```

```python
def get_director(x):
  l=[]
  for item in ast.literal_eval(x):
    if item['job']=='Director':
      l.append(item['name'])
      break
  return l
```

```python
movies_df['crew']=movies_df['crew'].apply(get_director)
```

```python
movies_df.head()
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[item.replace(' ','') for item in x])
```

```python
movies_df['genres']
```

```python
movies_df['crew']
```

```python
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[item.replace(' ','') for item in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[item.replace(' ','') for item in x])
```

```python
movies_df.isna().sum()
```

```python
movies_df['crew']=movies_df['crew'].apply(lambda x:[item.replace(' ','') for item in x])
```

```python
movies_df.head()
```

```python
movies_df['tags']=movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
```

```python
new_df=movies_df[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend_movies(movie):
  index=new_df[new_df['title']==movie].index[0]
  similarity_scores=similarity[index]
  similarity_scores=list(enumerate(similarity_scores))
  movies_list=sorted(similarity_scores,reverse=True,key=lambda x:x[1])[1:6]
  l=[]
  for movie in movies_list:
      l.append(new_df.iloc[movie[0]].title)

  return l
```

```python
recommend_movies('Avatar')
```
